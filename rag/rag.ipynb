```python
#文本多路召回 重排序
import jieba, json
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from rank_bm25 import BM25Okapi

# 加载重排序模型
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-reranker-large")
rerank_model = AutoModelForSequenceClassification.from_pretrained("BAAI/bge-reranker-large")
rerank_model.cuda()
rerank_model.eval()


# bge = json.load(open('/content/drive/MyDrive/iss/5004/submit_embedding_retrieval_top10.json'))
# bm25 = json.load(open('/content/drive/MyDrive/iss/5004/submit_bm25_retrieval_top10.json'))
```

```python
import chromadb

def get_context_collection():
  chroma_client = chromadb.PersistentClient(path="/content/drive/MyDrive/5004_rag/5004/chroma_output_e5")#路径别写空格
  collection = chroma_client.get_or_create_collection(name="question_embeddings_e5") # question
  collection2 = chroma_client.get_or_create_collection(name="pdf_embeddings_e5")   #
  return collection,collection2
```

```python
## 需要思考的一点 input和问题匹配。 input和context直接匹配 考虑一下数据库设计 如果与context分数更高直接选择
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel
from rank_bm25 import BM25Okapi
import jieba

from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel

# model = SentenceTransformer('../hugging-face-model/moka-ai/m3e-small/')
model = SentenceTransformer("intfloat/e5-large-v2")

# 加载重排序模型
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-reranker-large")
rerank_model = AutoModelForSequenceClassification.from_pretrained("BAAI/bge-reranker-large")
rerank_model.cuda()
rerank_model.eval()

def remove_duplicates(input_list):
  seen =set()
  result = []
  for item in input_list:
    if item not in seen:
      seen.add(item)
      result.append(item)
  return result

def match(input):
# 匹配问题寻找相关的上下文
  collection_q, collection_c = get_context_collection()

  #因为从collection拿到的文本没有id 要按id排序以后使用
  question = collection_q.get()
  df_q = pd.DataFrame(question)
  df_q['ids'] = df_q['ids'].apply(lambda x: int(x))
  df_q = df_q.sort_values(by='ids')
  df_q.reset_index(drop=True, inplace=True)


  # bm25
  content = [jieba.lcut(context) for context in df_q['documents'].tolist()]
  bm25 = BM25Okapi(content)
  doc_scores = bm25.get_scores(jieba.lcut(input)) # array
  bm25_index = doc_scores.argsort()[::-1][:3].astype(str)
  print(bm25_index)
  #判断一下是否需要加一

  # e5
  input_embedding = model.encode(input,normalize_embeddings=True).tolist()
  embedding_index = collection_q.query(
    query_embeddings=input_embedding,
    n_results=6)['ids'][0]
  print(embedding_index)

  #重排序
  fusion_result = []
  k = 60
  fusion_score = {}
  for idx,q in enumerate(bm25_index):#页数排名idx:页数
    if q not in fusion_score:
      fusion_score[q]=1/(idx+ k)
    else:
      fusion_score[q]+=1/(idx + k)

  for idx,q in enumerate(embedding_index):
    if q not in fusion_score:
      fusion_score[q]=1/(idx+ k)
    else:
      fusion_score[q]+=1/(idx + k)
  sorted_dict = sorted(fusion_score.items(),key=lambda item: item[1], reverse=True) #对两种方法得到的值进行排序

  pairs = []
  for sorted_result in sorted_dict[:5]:
      id = sorted_result[0] #index和实际的id相差为1
      # print(id)
      # print(sorted_result)
      #找到对应id的context
      context = collection2.get(ids = [id])['documents'][0]
      pairs.append([input,context])
  inputs =tokenizer(pairs,padding=True,truncation=True,return_tensors='pt',max_length = 512)#将问题和检索得到的内容进行拼接
  with torch.no_grad():
    inputs = {key: inputs[key].cuda() for key in inputs.keys()}
    scores = rerank_model(**inputs, return_dict=True).logits.view(-1, ).float()
  sorted_result = sorted_dict[scores.cpu().numpy().argmax()]
  match_context = collection2.get(ids = [sorted_result[0]])['documents'][0]

  return match_context


```
